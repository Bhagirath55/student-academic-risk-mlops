# EDA for Student Dataset

## 1. Load Dataset

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# laod the dataset
data = pd.read_csv("student_data.csv")
data.head()
```

```python
# Separating Features and target
features = data.drop(['id','Target'], axis=1)
target = data['Target']
```

# Data Information

# Features

```python
variables_train = [col for col in features.columns if features[col].nunique() <= 10]

for var in variables_train:

  print(f"====={var}=====")

  unique_values = features[var].unique()

  n_unique_values = features[var].nunique()

  value_counts = features[var].value_counts()



  print(f"Unique values in {var}: {unique_values}")

  print(f"Number of unique values in {var}: {n_unique_values}")

  print(f"Value counts in {var}:\n{value_counts}\n")
```

# EDA

# Features

```python
num_columns_train = [col for col in features if col not in ['id'] + variables_train]
fig, axes = plt.subplots(7,6, figsize = (20,16))
axes = axes.flatten()

for i, col in enumerate(num_columns_train):
  ax = sns.histplot(x = features[col], data = features, ax = axes[i], kde = True)
  ax.set_title(f'{col} Distribution')
  ax.set_xlabel(f'{col}')

for j in range(i+1, len(axes)):
  fig.delaxes(axes[j])

fig.suptitle('Distribution of Features per Dataset')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(len(variables_train), 2, figsize=(16, len(variables_train) * 5))

for i, column in enumerate(variables_train):
  ax_pie = axes[i, 0]
  values = features[column].value_counts()
  explode = [0.1] * len(values)

  ax_pie.pie(
        values,
        labels=values.index,
        autopct='%1.1f%%',
        shadow=True,
        explode=explode,
        startangle=140
  )
    
  ax_pie.set_title(f'{column} Distribution (Pie Chart)')

  ax_count = axes[i, 1]
  sns.countplot(
    y=column,
    data=features,
    order=values.index,
    ax=ax_count,
  )
    
  ax_count.set_xlabel('Count')
  ax_count.set_ylabel('')
  ax_count.set_title(f'{column} Distribution (Countplot)')

fig.suptitle('Distribution of Categorical Features in Train Dataset \n\n\n\n\n')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(7,6, figsize = (20,16))
axes = axes.flatten()

for i, col in enumerate(num_columns_train):
  ax = sns.boxplot(data = features[col], ax = axes[i])
  ax.set_title(f'{col} Distribution')
  ax.set_xlabel(f'{col}')

for j in range(i + 1, len(axes)):
  fig.delaxes(axes[j])

fig.suptitle('Distribution of Features per data')
plt.tight_layout()
plt.show()
```

```python
corr = features[num_columns_train].corr()
plt.figure(figsize=(20, 10))
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(corr, mask=mask, cmap='viridis', annot=True)
plt.title('Feature Dataset Correlation Matrix')
plt.show()
```
