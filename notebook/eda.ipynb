# EDA

# Features

```python
num_columns_train = [col for col in features if col not in ['id'] + variables_train]
fig, axes = plt.subplots(7,6, figsize = (20,16))
axes = axes.flatten()

for i, col in enumerate(num_columns_train):
  ax = sns.histplot(x = features[col], data = features, ax = axes[i], kde = True)
  ax.set_title(f'{col} Distribution')
  ax.set_xlabel(f'{col}')

for j in range(i+1, len(axes)):
  fig.delaxes(axes[j])

fig.suptitle('Distribution of Features per Dataset')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(len(variables_train), 2, figsize=(16, len(variables_train) * 5))

for i, column in enumerate(variables_train):
  ax_pie = axes[i, 0]
  values = features[column].value_counts()
  explode = [0.1] * len(values)

  ax_pie.pie(
        values,
        labels=values.index,
        autopct='%1.1f%%',
        shadow=True,
        explode=explode,
        startangle=140
  )
    
  ax_pie.set_title(f'{column} Distribution (Pie Chart)')

  ax_count = axes[i, 1]
  sns.countplot(
    y=column,
    data=features,
    order=values.index,
    ax=ax_count,
  )
    
  ax_count.set_xlabel('Count')
  ax_count.set_ylabel('')
  ax_count.set_title(f'{column} Distribution (Countplot)')

fig.suptitle('Distribution of Categorical Features in Train Dataset \n\n\n\n\n')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(7,6, figsize = (20,16))
axes = axes.flatten()

for i, col in enumerate(num_columns_train):
  ax = sns.boxplot(data = features[col], ax = axes[i])
  ax.set_title(f'{col} Distribution')
  ax.set_xlabel(f'{col}')

for j in range(i + 1, len(axes)):
  fig.delaxes(axes[j])

fig.suptitle('Distribution of Features per data')
plt.tight_layout()
plt.show()
```

```python
corr = features[num_columns_train].corr()
plt.figure(figsize=(20, 10))
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(corr, mask=mask, cmap='viridis', annot=True)
plt.title('Feature Dataset Correlation Matrix')
plt.show()
```


## 🔍 Feature Correlation Analysis and Feature Reduction

After performing correlation analysis between features and the target variable,
we identified several features with very weak or negative correlations (|corr| < 0.05)
that do not contribute meaningfully to the prediction of student academic risk.

### ❌ Removed Features:
- Educational special needs
- International
- Nationality
- Father's qualification
- Father's occupation
- Mother's occupation
- Mother's qualification
- Inflation rate
- Previous qualification
- Curricular units 1st sem (without evaluations)
- Curricular units 2nd sem (without evaluations)

Optionally removed:
- Marital status (low impact on target)

These features were dropped to improve model efficiency and reduce noise
without losing important predictive information.

The cleaned dataset is saved as `Desktop/AI-ML-Projects/student-academic-risk-mlops/data/preprocessed_data/cleaned_student_dataset.csv`.
